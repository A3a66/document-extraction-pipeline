```python
from openai import OpenAI

client = OpenAI()

def ask(prompt, system="You are a helpful assistant.", temp=0.7):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": prompt}
        ],
        temperature=temp
    )
    return response

print(ask("Explain prompt engineering simply"))
```
